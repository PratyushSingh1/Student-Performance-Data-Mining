```python
dataset['total'] = dataset['math score'] + dataset['reading score'] + dataset['writing score']
dataset['average'] = dataset['total']/3
dataset['average']
```

```python
dataset['math_passStatus'] = np.where(dataset['math score']<40, 'F', 'P')
dataset['read_passStatus'] = np.where(dataset['reading score']<40, 'F', 'P')
dataset['writing_passStatus'] = np.where(dataset['writing score']<40, 'F', 'P')
```
